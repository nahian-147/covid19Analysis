```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
```

```python
data = pd.read_csv('covid-data.csv')
```

```python
data
```

```python
countries = data['location'].unique()
```

```python
def DeathByCountries(country):
    %matplotlib inline
    
    cntryData = data[data['location']==country]
    cntryData = cntryData.dropna()
    ttlCase = cntryData['total_cases']
    nwCase = cntryData['new_cases']
    ttlDeath = cntryData[cntryData['total_deaths']>0].total_deaths
    nwDeath = cntryData['new_deaths']
    dateDth = cntryData[cntryData['total_deaths']>0].date
    date = cntryData['date']
    
    plt.figure(figsize=(55,20))
    
    plt.suptitle('Date vs Total Death and Daily New Death of '+str(country),fontsize = 55)
    
    plt.subplot(2, 1, 1)
    plt.bar(dateDth,ttlDeath,color=sns.color_palette('Greys',len(dateDth)),width=0.3)
    plt.yticks(fontsize=25)
    plt.xticks(rotation=90)
    plt.xlabel('Date',fontsize=40)
    plt.ylabel('Total Death',fontsize=40)
    
    plt.subplot(2, 1, 2)
    plt.bar(date, nwDeath,color='Grey',width=0.27)
    plt.xlabel('Date',fontsize=40)
    plt.ylabel('New Death',fontsize=40)
    plt.yticks(fontsize=25)
    plt.xticks(rotation=90)
```

```python
DeathByCountries('United States')
```

```python
Countries = widgets.Dropdown(
            options=countries,
            disabled=False,
            )
```

```python
widgets.interact(DeathByCountries,country=countries);
```

```python
def CaseByCountries(country):
    %matplotlib inline
    
    cntryData = data[data['location']==country]
    cntryData = cntryData.dropna()
    ttlCase = cntryData['total_cases']
    nwCase = cntryData['new_cases']
    date = cntryData['date']
    
    plt.figure(figsize=(55,20))
    
    plt.suptitle('Date vs Total Case and Daily New Case of '+str(country),fontsize = 55)
    
    plt.subplot(2, 1, 1)
    plt.bar(date,ttlCase,color=sns.color_palette('Greys',len(date)),width=0.3)
    plt.yticks(fontsize=25)
    plt.xticks(rotation=90)
    plt.xlabel('Date',fontsize=40)
    plt.ylabel('Total Case',fontsize=40)
    
    plt.subplot(2, 1, 2)
    plt.bar(date, nwCase,color='Grey',width=0.3)
    plt.xlabel('Date',fontsize=40)
    plt.ylabel('New Case',fontsize=40)
    plt.yticks(fontsize=25)
    plt.xticks(rotation=90)
```

```python
CaseByCountries('Bangladesh')
```

```python
def PlotCase_vs_Death(country):
    %matplotlib inline
    
    
    cntryData = data[data['location']==country]
    cntryData = cntryData.dropna()
    ttlCase = cntryData['total_cases']
    ttlDeath = cntryData['total_deaths']
    
    plt.figure(figsize=(30,10))
    plt.plot(ttlCase,ttlDeath,color='Grey')
    plt.plot(ttlCase,ttlDeath,'ko')
    plt.xlabel('Total Cases',fontsize = 35)
    plt.ylabel('Total Deaths',fontsize = 35)
    plt.xticks(fontsize = 25)
    plt.yticks(fontsize = 25)
    plt.title('Total Case vs Total Deaths of '+str(country),fontsize = 40)
```

```python
PlotCase_vs_Death('Bangladesh')
```

```python
def PlotDteTtlCsDth(country):
    %matplotlib inline
    
    cntryData = data[data['location']==country]
    cntryData = cntryData.dropna()
    ttlCase = cntryData['total_cases']
    ttlDeath = cntryData['total_deaths']
    date = cntryData['date']
    
    plt.figure(figsize=(30,10))
    plt.suptitle('Date vs Total Case and Total Death of '+str(country),fontsize = 57)
    
    plt.subplot(2, 1, 1)
    plt.plot(date,ttlCase,color='Grey')
    plt.plot(date,ttlCase,'ko')
    plt.xlabel('Date',fontsize = 40)
    plt.ylabel('Total Case',fontsize = 40)
    plt.xticks(fontsize =10,rotation=90)
    plt.yticks(fontsize = 25)
    
    plt.subplot(2, 1, 2)
    plt.plot(date,ttlDeath,'grey')
    plt.plot(date,ttlDeath,'ko')
    plt.xlabel('Date',fontsize = 40)
    plt.ylabel('Total Death',fontsize = 40)
    plt.xticks(fontsize =10,rotation=90)
    plt.yticks(fontsize = 25)
```

```python
def PlotTtlCase_vs_NwCase(country):
    %matplotlib inline
    
    cntryData = data[data['location']==country]
    cntryData = cntryData.dropna()
    ttlCase = cntryData['total_cases']
    nwCase = cntryData['new_cases']
    
    plt.figure(figsize=(30,10))
    plt.plot(ttlCase,nwCase,color='Grey')
    plt.plot(ttlCase,nwCase,'ko')
    plt.xlabel('Total Cases',fontsize = 35)
    plt.ylabel('New Cases',fontsize = 35)
    plt.xticks(fontsize = 25)
    plt.yticks(fontsize = 25)
    plt.title('Total Case vs New Case of '+str(country),fontsize = 45)
```

```python
Functions = widgets.Dropdown(
            options=[('Total Case vs New Case',PlotTtlCase_vs_NwCase),
                     ('Total Case vs Total Death',PlotCase_vs_Death),
                     ('Total Case and Daily New Case',CaseByCountries),
                     ('Total Death and Daily New Death',DeathByCountries),
                     ('Date vs Total Case and Total Death',PlotDteTtlCsDth)],
            disabled=False,
            )
```

```python
def myPlot(country,PlotFun):
    PlotFun(country)
```

```python
widgets.interact(myPlot,country=countries,PlotFun=Functions);
```
